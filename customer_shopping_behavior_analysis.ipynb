```python
# Loading the dataset using pandas
import pandas as pd

df = pd.read_csv('shopping_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Checking if duplicate data is present in the dataset

df.duplicated().sum()
```

```python
# Checking if missing data or null values are present in the dataset

df.isnull().sum()
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount_usd'})
```

```python
df.columns
```

```python
# create a new column age_group

labels = ['Young Adult','Adult','Middle Aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied']== df[ 'promo_code_used']).all()
```

```python
# Dropping promo_code_used column

df = df.drop('promo_code_used',axis=1)
```

```python
df.columns
```

```python
# Rearranging the columns

new_order = [
    'customer_id', 'age', 'age_group', 'gender', 'location',
    'item_purchased', 'category', 'size', 'color', 'season',
    'purchase_amount_usd', 'discount_applied', 'payment_method', 'shipping_type',
    'previous_purchases', 'frequency_of_purchases', 'purchase_frequency_days',
    'subscription_status', 'review_rating'
]


df = df[new_order]
```
